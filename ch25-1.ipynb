```python
(fig, ax)=plt.subplots()

x=[2013, 2014, 2015, 2016, 2017, 2018, 2019]
y1=[1.6, 1.5, 2.9, 2.8, 2.3, 3.6, 2.6]
y2=[3.1, 3.1, 2.8, 2.9, 3.1, 2.9, 2.2]

ax.plot(x, y1, linestyle='--', marker='o',
       markerfacecolor='w', markeredgecolor='r')
ax.plot(x, y2, linestyle=':', color='gray', 
        marker='D',markerfacecolor='w', 
        markeredgecolor='k')
```

```python
(fig, ax)=plt.subplots()

x=[2013, 2014, 2015, 2016, 2017, 2018, 2019]
y1=[1.6, 1.5, 2.9, 2.8, 2.3, 3.6, 2.6]
y2=[3.1, 3.1, 2.8, 2.9, 3.1, 2.9, 2.2]

ax.plot(x, y1, linestyle='--', marker='o',
       markerfacecolor='w', markeredgecolor='r',
       label='Seoul')
ax.plot(x, y2, linestyle=':', color='gray', 
        marker='D',markerfacecolor='w', 
        markeredgecolor='k', label='nationwide')

ax.set_title('Rate of Economic Growth')
ax.set_xlabel('Year')
ax.set_ylabel('Growth Rate')
ax.legend(loc='lower right')
```

```python
ax.set_ylim(0, 4)
```

```python
ax.set_ylim(0, 4)
ax.set_yticks([0, 1.5, 2.0, 2.5, 3.0, 3.5])
```

```python
ax.set_xticks(range(2013, 2020))
ax.set_xticklabels(['Base Year\n(B)', 'B+1', 'B+2', 'B+3', 'B+4', 'B+5', 'B+6'])
```

```python
import numpy as np

fig, [ax1, ax2, ax3]=plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

x1=[1, 2, 3, 4]
y1=[10, 9, 3, 5]
ax1.plot(x1, y1, marker='o')

x2=np.arange(-1, 2, 0.1)
y2=[x**2 for x in x2]
ax2.plot(x2, y2)

x3=['A-Mart', 'B-Mart', 'C-Mart']
y3=[100, 63, 85]
ax3.bar(x3, y3, width=0.4)
```

```python
import numpy as np

fig, [ax1, ax2, ax3]=plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

x1=[-1, 0, 1, 2]
y1=[x**2 for x in x1]
ax1.plot(x1, y1, marker='o')

x2=np.arange(-1, 2, 0.5)
y2=[x**2 for x in x2]
ax2.plot(x2, y2, marker='o')

x3=np.arange(-1, 2, 0.1)
y3=[x**2 for x in x3]
ax3.plot(x3, y3, marker='o')
```
